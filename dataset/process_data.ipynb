```python
import pandas as pd
import numpy as np
import os
```

```python
data_path = "/media/jionie/my_disk/Kaggle/Bengaliai/input/bengaliai-cv19"
train_df = pd.read_csv(data_path + '/train.csv')
```

```python
train_df.head()
```

```python
class_map = pd.DataFrame() 
class_map['grapheme'] = train_df['grapheme'].unique()
class_map['label'] = np.array([i for i in range(len(class_map))])
```

```python
grapheme_list = list(class_map['grapheme'])
```

```python
grapheme_root_list = []
vowel_diacritic_list = []
consonant_diacritic_list = []
for grapheme in grapheme_list:
    grapheme_root = train_df[train_df['grapheme'] == grapheme]['grapheme_root'].values[0]
    vowel_diacritic = train_df[train_df['grapheme'] == grapheme]['vowel_diacritic'].values[0]
    consonant_diacritic = train_df[train_df['grapheme'] == grapheme]['consonant_diacritic'].values[0]
    
    grapheme_root_list.append(grapheme_root)
    vowel_diacritic_list.append(vowel_diacritic)
    consonant_diacritic_list.append(consonant_diacritic)
```

```python
class_map['grapheme_root'] = np.array(grapheme_root_list)
class_map['vowel_diacritic'] = np.array(vowel_diacritic_list)
class_map['consonant_diacritic'] = np.array(consonant_diacritic_list)
```

```python
class_map.to_csv("grapheme_1295.csv")
```
