```python
import pandas as pd
import numpy as np
import os
```

```python
data_path = "/media/jionie/my_disk/Kaggle/Bengaliai/input/bengaliai-cv19"
train_df = pd.read_csv(data_path + '/train.csv')
```

```python
train_df.head()
```

```python
class_map = pd.DataFrame() 
class_map['grapheme'] = train_df['grapheme'].unique()
class_map['label'] = np.array([i for i in range(len(class_map))])
```

```python
grapheme_list = list(class_map['grapheme'])
```

```python
grapheme_root_list = []
vowel_diacritic_list = []
consonant_diacritic_list = []
for grapheme in grapheme_list:
    grapheme_root = train_df[train_df['grapheme'] == grapheme]['grapheme_root'].values[0]
    vowel_diacritic = train_df[train_df['grapheme'] == grapheme]['vowel_diacritic'].values[0]
    consonant_diacritic = train_df[train_df['grapheme'] == grapheme]['consonant_diacritic'].values[0]
    
    grapheme_root_list.append(grapheme_root)
    vowel_diacritic_list.append(vowel_diacritic)
    consonant_diacritic_list.append(consonant_diacritic)
```

```python
class_map['grapheme_root'] = np.array(grapheme_root_list)
class_map['vowel_diacritic'] = np.array(vowel_diacritic_list)
class_map['consonant_diacritic'] = np.array(consonant_diacritic_list)
```

```python
class_map.to_csv("grapheme_1295.csv")
```

```python
def decode(prediction, class_map):
    
    p0, p1, p2, p3 = prediction[0], prediction[1], prediction[2], prediction[3]
    
    for b in range(p0.shape[0]):
        
        p0[b] = class_map[class_map['label'] == p3[b]]['grapheme_root'].values[0]
        p1[b] = class_map[class_map['label'] == p3[b]]['vowel_diacritic'].values[0]
        p2[b] = class_map[class_map['label'] == p3[b]]['consonant_diacritic'].values[0]
        
    
    return p0,p1,p2,p3

grapheme_root_prediction = np.random.randint(168, size=100)
vowel_diacritic_prediction = np.random.randint(11, size=100)
consonant_diacritic_prediction = np.random.randint(7, size=100)
grapheme_prediction = np.random.randint(1295, size=100)

grapheme_root_prediction_decode, \
vowel_diacritic_prediction_decode, \
consonant_diacritic_prediction_decode, \
grapheme_prediction_decode = decode([grapheme_root_prediction.copy(), vowel_diacritic_prediction.copy(), \
                            consonant_diacritic_prediction.copy(), grapheme_prediction.copy()], class_map)
```
